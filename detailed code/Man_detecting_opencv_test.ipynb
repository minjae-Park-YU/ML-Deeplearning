```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
import time
```

```python
for i in range(25):
    image_name = './CCTV_dataset/train/theft/video_2076224.MP4_00/frame_04' + str(590 + 15*i) + '.jpg'
    print(image_name)
    image = cv2.imread(image_name)
    fig = plt.figure(1)
    ax1 = fig.add_subplot(1, 2, 1)
    ax2 = fig.add_subplot(1, 2, 2)
    grayImage = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    
    body_cascade = cv2.CascadeClassifier('./haarcascades/haarcascade_fullbody.xml')
    body = body_cascade.detectMultiScale(grayImage, 1.01, 5)

    for (x,y,w,h) in body:
        cv2.rectangle(image, (x,y), (x+w, y+h), (0,0,255), 3)
            
    ax2.imshow(image)
    ax1.imshow(grayImage)
    plt.show()
```
